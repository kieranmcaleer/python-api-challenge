### Humidity Heatmap
* Configure gmaps.
* Use the Lat and Lng as locations and Humidity as the weight.
* Add Heatmap layer to map.

```python
#configuring gmaps
gmaps.configure(api_key=gkey)
```

```python
#create the locations using lng and lat for use in the heat layer
locations = cleaned_df[['Latitude', 'Longitude']]
#set the weight to be humidity
weights= cleaned_df["humidity"]
#create the heat map layer with the values we just created
heat_layer = gmaps.heatmap_layer(locations, weights=weights, point_radius = 15, max_intensity=100)
#make a figure and put the heat map layer on it
fig = gmaps.figure()
fig.add_layer(heat_layer)
fig
```

### Create new DataFrame fitting weather criteria

```python
#create a new df with specifications for vacation
vacation_df = cleaned_df.loc[(cleaned_df["max_temp"]>=70)&(cleaned_df["max_temp"]<=90)]
vacation_df = vacation_df.loc[vacation_df["Cloudiness"]<=5]
vacation_df = vacation_df.loc[vacation_df["Wind Speed"]<=8]
vacation_df= vacation_df.dropna()
vacation_df
```

### Hotel Map
* Store into variable named `hotel_df`.
* Add a "Hotel Name" column to the DataFrame.
* Set parameters to search for hotels with 5000 meters.
* Hit the Google Places API for each city's coordinates.
* Store the first Hotel result into the DataFrame.
* Plot markers on top of the heatmap.

```python
#make that equal to a new dataframe for hotels
hotel_df=vacation_df
hotel_df
```

```python
#https://maps.googleapis.com/maps/api/place/nearbysearch/json?location=-33.8670522,151.1957362&radius=1500&type=restaurant&keyword=cruise&key=YOUR_API_KEY
#the link above is what we are trying to create
#set all of the parameters 
key=gkey
target_radius = 5000
#there is no place type hotels so lodging is the best option
target_type = "lodging"
base_url = "https://maps.googleapis.com/maps/api/place/nearbysearch/json"
#put all the params in a dictionary so that we can use that for later in the url
params={
    "radius":target_radius,
    "type":target_type,
    "key":key,
    "keyword":"hotel"
}
#going through the rows of the dataframe to find information for out search
try:
    for index, row in hotel_df.iterrows():
        #save the lat and long so that we can make it a param
        lat=row["Latitude"]
        long=row["Longitude"]
        params["location"] = f"{lat},{long}"
    #get the api request
        response = requests.get(base_url, params=params).json()
        #set the hotel name in the dataframe equal to the one that pops up in the search
        hotel_df.loc[index, "Hotel Name"]= response["results"][0]["name"]
        #print out the closest hotel to each of our cities
        print(f"The closest hotel to {row['Cities']} is {response['results'][0]['name']} ")
except IndexError:
    print("Skipping because of error")

    
```
